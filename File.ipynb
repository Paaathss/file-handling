The key for working with files with python is the open()function .
The open() takes two parameters ::: filename , mode
There are 4 diff methods(modes) for opening a file:

```python
open('filename','mode')
```

Mode description

"r"--Read-- Default value. Opens file for reading , error if file does not exists .

"w"--writing-- This mode open file for writing , if file doesn't exists ,it creates a new file  , if file existes it truncates the file .

"x"--creates a new file-- if file already exists, the operation fails .

"a"--Open file in append mode-- if file does not exists,it creates a new file .

"t"--default mode-- it opens in text mode .

"b"--it opens in binary mode .

"+"--This will open a file for reading and writing(updating)
